```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('population.csv')
df.head()
```

```python
df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
chosen_country = 'India'

country_data = df_final[df_final['Country Name'] == chosen_country]

years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

#BAR CHART
plt.figure(figsize=(8, 5))
plt.bar(years, population, color='steelblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y')

for i, (year, pop) in enumerate(zip(years, population)):
    if i % 10 == 0:
        plt.text(year, pop, f'{pop:,}', ha='center', va='bottom', fontsize=8, rotation=30)

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.barh(years, population, color='steelblue')
plt.xlabel('Population')
plt.ylabel('Year')
plt.title(f'Population of {chosen_country} Over the Years')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')

plt.show()
```

```python
#HISTOGRAM
plt.figure(figsize=(8, 5))
plt.hist(population, bins=20, color='steelblue', edgecolor='black', alpha=0.7)

plt.title('Population Distribution Histogram')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')

plt.show()
```

```python
country1 = df_final[df_final['Country Name'] == 'India'].iloc[:, 2:].values.flatten()
country2 = df_final[df_final['Country Name'] == 'China'].iloc[:, 2:].values.flatten()

plt.figure(figsize=(8, 5))

plt.hist(country1, bins=30, color='steelblue', edgecolor='black', alpha=0.6, label='India',density=True)
plt.hist(country2, bins=30, color='salmon', edgecolor='black', alpha=0.6, label='China',density=True)

plt.title('Population Distribution Comparison')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
